```python
def calculate(num1, num2, operator):
    if not isinstance(num1, (int, float)) or not isinstance(num2, (int, float)):
        print("Ошибка: оба параметра должны быть числами")
        return
    if operator == "+":
        return num1 + num2
    elif operator == "-":
        return num1 - num2
    elif operator == "*":
        return num1 * num2
    elif operator == "/":
        if num2 == 0:
            print ("Ошибка: деление на ноль!")
        return num1 / num2
    else:
        print (f"Ошибка: недопустимый оператор '{operator}'")

```

```python
calculate(5,5, "+")
```
