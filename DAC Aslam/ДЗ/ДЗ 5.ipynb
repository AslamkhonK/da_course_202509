```python
n=int(input('введите число:' ))

numbers=list(range (1,n+1))
print (f'числа от 1 до {n}:')
print (numbers)
```

```python
num = int(input("Введите число: "))

divisors = [i for i in range(1, num + 1) if num % i == 0]
print(f"Список делителей числа {num}:")
print(divisors)
```

```python
ip = input("Введите IP-адрес: ")

parts = ip.split(".")

if len(parts) == 4 and all(p.isdigit() and 0 <= int(p) <= 255 for p in parts):
    print("Корректно")
else:
    print("Не корректно")
```

```python
text = input("Введите строку: ")

numbers = [int(ch) for ch in text if ch.isdigit()]

print("Список чисел из строки:")
print(numbers)
```

```python
lst_numbers = [7, 3, -4, 1, 0, -1, 0, -2]

lst_pos = [x for x in lst_numbers if x > 0]  
lst_neg = [x for x in lst_numbers if x < 0]   
lst_zrs = [x for x in lst_numbers if x == 0]  

print("lst_pos -", lst_pos)
print("lst_zrs -", lst_zrs)
print("lst_neg -", lst_neg)
```
