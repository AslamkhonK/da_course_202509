```python
class Scales:
    def __init__(self):
        self.left = 0
        self.right = 0

    def add_left(self, weight):
        self.left += weight

    def add_right(self, weight):
        self.right += weight

    def get_result(self):
        if self.left == self.right:
            return "Весы в равновесии"
        elif self.right > self.left:
            return "Правая чаша тяжелее"
        else:
            return "Левая чаша тяжелее"



```

```python

scales = Scales()

scales.add_right(1)
scales.add_right(2)
scales.add_left(2)

print(scales.get_result())
```

```python
class Todo:
    def __init__(self):
        self.things = []

    def add(self, name, priority):
        self.things.append((name, priority))

    def get_by_priority(self, n):
        return [name for name, p in self.things if p == n]

    def get_low_priority(self):
        if not self.things:
            return []
        min_priority = min(p for _, p in self.things)
        return [name for name, p in self.things if p == min_priority]

    def get_high_priority(self):
        if not self.things:
            return []
        max_priority = max(p for _, p in self.things)
        return [name for name, p in self.things if p == max_priority]


```

```python
todo = Todo()

print(todo.things)               # []
print(todo.get_by_priority(1))   # []
print(todo.get_low_priority())   # []
print(todo.get_high_priority())  # []

```

```python
todo.add("Помыть посуду", 2)
todo.add("Купить продукты", 1)
todo.add("Сделать ДЗ", 3)
todo.add("Погулять", 1)

print(todo.things)


print(todo.get_by_priority(1))


print(todo.get_low_priority())


print(todo.get_high_priority())
```

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def get_balance(self):
        return self.balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            raise ValueError("На счете недостаточно средств")
        self.balance -= amount

    def transfer(self, account, amount):
        if amount > self.balance:
            raise ValueError("На счете недостаточно средств")
        self.balance -= amount
        account.balance += amount
```

```python

account = BankAccount()

print(account.get_balance()) 
account.deposit(100)
print(account.get_balance())  
account.withdraw(50)
print(account.get_balance()) 

account2 = BankAccount(200)
account.transfer(account2, 25)

print(account.get_balance())   
print(account2.get_balance())  
```

```python
account1 = BankAccount(100)
account2 = BankAccount(200)

account1.transfer(account2, 50)
print(account1.get_balance())
print(account2.get_balance())
```

```python
account = BankAccount(100)

try:
    account.withdraw(150)
except ValueError as e:
    print(e)
```
