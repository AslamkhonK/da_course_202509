```python
class PhoneNumber:
    def __init__(self, phone_number: str):
    
        self.number = phone_number.replace(" ", "")
        if len(self.number) != 10 or not self.number.isdigit():
            raise ValueError("Номер должен содержать ровно 10 цифр")
    
    def __str__(self):
        
        return f"({self.number[:3]}) {self.number[3:6]}-{self.number[6:]}"
    
    def __repr__(self):
        return f"PhoneNumber('{self.number}')"


```

```python
phone1 = PhoneNumber('9173963385')
phone2 = PhoneNumber('918 396 3389')
print(phone1, phone2, sep=", ")
print([phone1, phone2])

```

```python

class Month:
    def __init__(self, year: int, month: int):
        self.year = year
        self.month = month
    
    def __str__(self):
        return f"{self.year}-{self.month}"
    
    def __repr__(self):
        return f"Month({self.year}, {self.month})"
    
   
    def __eq__(self, other):
        if not isinstance(other, Month):
            return NotImplemented
        return (self.year, self.month) == (other.year, other.month)
    
    def __lt__(self, other):
        if not isinstance(other, Month):
            return NotImplemented
        return (self.year, self.month) < (other.year, other.month)
    
    def __le__(self, other):
        if not isinstance(other, Month):
            return NotImplemented
        return (self.year, self.month) <= (other.year, other.month)
    
    def __gt__(self, other):
        if not isinstance(other, Month):
            return NotImplemented
        return (self.year, self.month) > (other.year, other.month)
    
    def __ge__(self, other):
        if not isinstance(other, Month):
            return NotImplemented
        return (self.year, self.month) >= (other.year, other.month)
    
    def __ne__(self, other):
        if not isinstance(other, Month):
            return NotImplemented
        return (self.year, self.month) != (other.year, other.month)


```

```python
months = [Month(1998, 12), Month(2000, 1), Month(1999, 12)]
print(sorted(months))
print(min(months))
print(max(months))
```

```python

class Version:
    def __init__(self, version: str):
        parts = version.split(".")
        
        while len(parts) < 3:
            parts.append("0")
        self.numbers = tuple(int(p) for p in parts)
        self.version_str = ".".join(str(n) for n in self.numbers)
    
    def __str__(self):
        return self.version_str
    
    def __repr__(self):
        return f"Version('{self.version_str}')"
    
    def __eq__(self, other):
        if not isinstance(other, Version):
            return NotImplemented
        return self.numbers == other.numbers
    
    def __lt__(self, other):
        if not isinstance(other, Version):
            return NotImplemented
        return self.numbers < other.numbers
    
    def __le__(self, other):
        if not isinstance(other, Version):
            return NotImplemented
        return self.numbers <= other.numbers
    
    def __gt__(self, other):
        if not isinstance(other, Version):
            return NotImplemented
        return self.numbers > other.numbers
    
    def __ge__(self, other):
        if not isinstance(other, Version):
            return NotImplemented
        return self.numbers >= other.numbers
    
    def __ne__(self, other):
        if not isinstance(other, Version):
            return NotImplemented
        return self.numbers != other.numbers
```

```python
versions = [Version('2'), Version('2.1'), Version('1.9.1')]
print(sorted(versions))
print(min(versions))
print(max(versions))
```
