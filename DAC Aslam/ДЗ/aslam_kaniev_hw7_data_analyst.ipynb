```python
 #1.У случайной выборки людей разного возраста спросили надевают ли они очки при чтении.
 # Результаты приведены в таблице ниже. Имеется ли взаимосвязь между возрастом и ношением очков?


from scipy.stats import chi2_contingency
observed = np.array([
    [15, 56],
    [19, 37],
    [34, 35],
    [51, 23]
])

chi2, p, dof, expected = chi2_contingency(observed)

print("Chi-square:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)

alpha = 0.05
if p < alpha:
    print("Вывод: существует связь между возрастом и ношением очков.")
else:
    print("Вывод: нет оснований утверждать, что переменные связаны.")
```

```python
#Случайная выборка студентов была разделена на 3 группы и каждая из них готовилась к сдаче экзамена по одной из 3 методик (A, B, C). В таблице ниже приведены оценки по экзамену студентов каждой группы.
#  Определите имеются ли различия между группами.

from scipy.stats import f_oneway
group1 = np.array([85, 86, 88, 75, 78, 94, 98, 79, 71, 80])
group2 = np.array([91, 92, 93, 85, 87, 84, 82, 88, 95, 96])
group3 = np.array([79, 78, 88, 94, 92, 85, 83, 85, 82, 81])

F, p = f_oneway(group1, group2, group3)

print("F-statistic:", F)
print("p-value:", p)

alpha = 0.05
if p < alpha:
    print("Вывод: группы отличаются статистически значимо.")
else:
    print("Вывод: различий между группами не обнаружено.")
```

```python
#3.Случайная выборка 20 игроков в баскетбол была отобрана для улучшения высоты их вертикального прыжка специальным методом тренировок. 
# Высота максимального вертикального прыжка (в сантиметрах) каждого игрока измерили при отборе а потом измерили повторно, после прохождения тренировок. 
# Результаты представлены в таблице ниже. Улучшила ли программа высоту вертикального прыжка игроков?




from scipy.stats import ttest_rel

before = np.array([
    55.88, 50.80, 48.26, 60.96, 63.50, 63.50, 71.12, 55.88, 76.20, 68.58,
    60.96, 45.72, 40.64, 48.26, 48.26, 71.12, 60.96, 63.50, 63.50, 58.42
])

after = np.array([
    60.96, 55.88, 48.26, 55.88, 71.12, 66.04, 71.12, 60.96, 76.20, 73.66,
    63.50, 50.80, 43.18, 45.72, 45.72, 71.12, 66.04, 68.58, 68.58, 60.96
])

t_stat, p = ttest_rel(after, before)

print("t-statistic:", t_stat)
print("p-value:", p)

alpha = 0.05
if p < alpha:
    print("Вывод: тренировочная программа улучшила среднюю высоту прыжка.")
else:
    print("Вывод: улучшений не обнаружено.")
```
