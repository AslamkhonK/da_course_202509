```python
def convert_to(values, type_to=int):
    return list(map(type_to, values))
numbers = [116, -411, 448, 636, -254, -295, 220, 216, 187, -150]
print(convert_to(numbers, str))
print(convert_to(numbers, float))
```

```python
rooms = [
    ("Переговорная", "OK", "OK", "OK"),
    ("Серверная", "OK", "FAIL", "OK"),
    ("Склад", "FAIL", "OK", "OK"),
    ("Кухня", "OK", "OK", "OK"),
    ("Хол", "OK", "OK", "FAIL"),
    ("Уборная", "OK", "OK", "OK"),
]

result = [(name, "Безопасна" if all(status == "OK" for status in sensors) else "Небезопасна")
          for name, *sensors in rooms]

print(result)
```
