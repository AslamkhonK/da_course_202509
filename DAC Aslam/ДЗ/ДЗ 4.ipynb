```python
s = input()

if "f" not in s:
    print("NO")
else:
    first = s.find("f") 
    last = s.rfind("f")   
    if first == last:
        print(first)      
    else:
        print(first, last) 
```

```python
string = input()

max_count = 0
result = ""

for ch in string:
    cnt = string.count(ch)
    if cnt > max_count or (cnt == max_count and ch >= result):
        max_count = cnt
        result = ch

print(result)
```
