```python
import random
player_hp = 100
computer_hp = 100
round_num = 1
while player_hp > 0 and computer_hp > 0:
    print(f"\nРаунд {round_num}:")
    damage_player = random.randint(10, 30)
    computer_hp -= damage_player
    if computer_hp < 0:
        computer_hp = 0
    print(f"Игрок атаковал и нанес {damage_player} урона. У компьютера осталось {computer_hp} здоровья.")
    if computer_hp <= 0:
        break
    damage_computer = random.randint(5, 25)
    player_hp -= damage_computer
    if player_hp < 0:
        player_hp = 0
    print(f"Компьютер атаковал и нанес {damage_computer} урона. У игрока осталось {player_hp} здоровья.")
    round_num += 1
print("\nИгра окончена!")
if player_hp > 0 and computer_hp <= 0:
    print("Победил игрок!")
elif computer_hp > 0 and player_hp <= 0:
    print("Победил компьютер!")
else:
    print("Ничья!")
```
