```python
import math

# Задача 1
# Четыре монеты подбрасываются разом.
# X — количество орлов. Найти P(X = 2).

n = 4        # число подбрасываний
k = 2        # количество орлов
p = 0.5      # вероятность орла в одном подбрасывании

# биномиальная формула: C(n, k) * p^k * (1 - p)^(n - k)
prob_task1 = math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

print("P(X = 2) =", prob_task1)
```

```python
# Задача 2
# Четыре монеты подбрасываются разом.
# X — количество орлов. Найти P(X ≥ 1).

n = 4
p = 0.5

# P(X >= 1) = 1 - P(X = 0)
k0 = 0
prob_x0 = math.comb(n, k0) * (p ** k0) * ((1 - p) ** (n - k0))
prob_task2 = 1 - prob_x0

print("P(X >= 1) =", prob_task2)
```

```python
# Задача 3
# Вероятность группы крови A равна 0.4.
# X — количество людей с группой A среди 4 случайно выбранных.
# Нужно построить биномиальное распределение P(X = x) для x = 0..4.

n = 4
p = 0.4

print("Биномиальное распределение X ~ Bin(n=4, p=0.4):")
for x in range(n + 1):
    prob_x = math.comb(n, x) * (p ** x) * ((1 - p) ** (n - x))
    print(f"P(X = {x}) = {prob_x:.5f}")
```

```python
# Задача 4
# Среднее число продаж в час λ = 10 (распределение Пуассона).
# Найти P(X = 8).

lam = 10      # λ
k = 8         # количество продаж

prob_task4 = math.exp(-lam) * (lam ** k) / math.factorial(k)

print("P(X = 8) =", prob_task4)
print("Округлённо:", round(prob_task4, 5))
```

```python
# Задача 5
# Игрок забивает штрафной с вероятностью p = 0.6.
# Он делает n = 12 бросков.
# Найти P(X = 10) — попадёт ровно 10 раз.

n = 12
k = 10
p = 0.6

prob_task5 = math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

print("P(X = 10) =", prob_task5)
print("Округлённо:", round(prob_task5, 5))
```
