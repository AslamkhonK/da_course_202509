```python
def get_days(month):
    if month in [1, 3, 5, 7, 8, 10, 12]:
        print(31)
    elif month in [4, 6, 9, 11]:
        print(30)
    else:
        print(28)  
```

```python
get_days(5)  
get_days(10)  
get_days(7)
```

```python
def words_length(words):
    return [len(word) for word in words]

print(words_length(['Hello', 'world!']))      
print(words_length(['Python', 'is', 'awesome!']))
print(words_length(['I', 'love', 'codiiiing']))  
```

```python
def is_password_good(password):
    
    if len(password) < 8:
        return False
    
    
    has_upper = any(char.isupper() for char in password)
    
    
    has_lower = any(char.islower() for char in password)
    
    
    has_digit = any(char.isdigit() for char in password)
    
    
    return has_upper and has_lower and has_digit


print(is_password_good('a#abbCC11OP'))  # True
print(is_password_good('CCCCu'))      # False
```

```python
def find_all(target, symbol):
    positions = []
    for i, char in enumerate(target):
        if char == symbol:
            positions.append(i)
    return positions


print(find_all('abcdabcaaa', 'd'))
print(find_all('abcadbcaaa', 'b'))  
print(find_all('abcadbcaaa', 'a')) 
```
