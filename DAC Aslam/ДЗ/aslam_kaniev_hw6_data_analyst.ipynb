```python
import math

def sample_mean(data):
    """Выборочное среднее."""
    return sum(data) / len(data)

def sample_std(data):
    """Выборочное стандартное отклонение (ddof = 1)."""
    m = sample_mean(data)
    n = len(data)
    return math.sqrt(sum((x - m) ** 2 for x in data) / (n - 1))

def binom_pmf(n, k, p):
    """Биномиальное распределение: P(X = k), X ~ Bin(n, p)."""
    return math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

def normal_ci_for_mean(x_bar, s, n, z):
    """Доверительный интервал для среднего при известном/большом n (нормальное приближение)."""
    se = s / math.sqrt(n)
    margin = z * se
    return x_bar - margin, x_bar + margin
```

## Задача 1

Имеется выборка возрастов (в годах):

\[
[6, 8, 11, 11, 17, 21, 34, 35, 41, 42, 43, 46, 54, 54, 59, 62]
\]

Каждое наблюдение — возраст одного человека.  
Найти точечные оценки:

- среднего значения популяции,
- стандартного отклонения популяции.


```python
data = [6, 8, 11, 11, 17, 21, 34, 35, 41, 42, 43, 46, 54, 54, 59, 62]

mu_hat = sample_mean(data)
sigma_hat = sample_std(data)

print("Точечная оценка среднего (μ̂) =", mu_hat)
print("Точечная оценка стандартного отклонения (σ̂) ≈", sigma_hat)
```

## Задача 2

24% всех M&M's имеют синий цвет. В стандартном пакете чаще всего 17 драже.  
Пусть X — количество синих драже в пакете, X ~ Bin(n = 17, p = 0.24).


```python
n = 17
p_blue = 0.24


prob = 0.0
for k in range(7, n + 1):
    prob += binom_pmf(n, k, p_blue)

print("P(доля синих > 40%) = P(X >= 7) ≈", prob)
```

## Задача 3

64 мужчины были выбраны случайным образом, их веса измерены.

- Среднее значение выборки: \(\bar{x} = 84\) кг
- Стандартное отклонение выборки: \(s = 24\) кг
- Объём выборки: \(n = 64\)

Найти **90% доверительный интервал** для среднего значения веса популяции.


```python
x_bar = 84
s = 24
n = 64
z_90 = 1.645 

ci_90 = normal_ci_for_mean(x_bar, s, n, z_90)

print("90% доверительный интервал для среднего веса:", ci_90)
```

##Задача 4

**В 2003 году Дартмутская Медицинская Школа запустила исследование для изучения связи между просмотром фильмов с курящими актёрами и началом курения среди подростков. В исследовании были случайным образом отобраны 6522 американских подростка в возрасте 10-14 лет, которые никогда не пробовали курить до этого. Из тех, кто впоследствии попробовал курить впервые, 38% сделали это из-за воздействия курения в фильмах. Оцените долю всех американских подростков в возрасте 10-14 лет, которые начали курить из-за просмотра курящих актеров в фильмах, построив 95% доверительный интервал.**

```python
n = 6522
p_hat = 0.38
z_95 = 1.96

se_p = math.sqrt(p_hat * (1 - p_hat) / n)
margin_p = z_95 * se_p

ci_p_95 = (p_hat - margin_p, p_hat + margin_p)

print("95% доверительный интервал для доли p:", ci_p_95)
print("В процентах: ({:.2f}%, {:.2f}%)".format(ci_p_95[0]*100, ci_p_95[1]*100))
```

**49 женщин, были выбраны случайным образом и их высота была измерена. Среднее значение выборки составило 164 см, стандартное отклонение - 20 см. Рассчитайте доверительный интервал для среднего значения популяции с 95%-м уровнем доверия.**

```python
x_bar = 164
s = 20
n = 49
z_95 = 1.96

ci_95 = normal_ci_for_mean(x_bar, s, n, z_95)

print("95% доверительный интервал для среднего роста:", ci_95)
```
