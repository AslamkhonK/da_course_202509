```python
class Rectangle:
    def __init__(self, width=1, height=1):
        self._validate(width)
        self._validate(height)
        self._width = width
        self._height = height

    def _validate(self, value):
        if not isinstance(value, (int, float)) or not (1 <= value <= 100):
            raise ValueError("Не применимо")

    def perimeter(self):
        return 2 * (self._width + self._height)

    def area(self):
        return self._width * self._height

    def get_width(self):
        return self._width

    def set_width(self, new_width):
        self._validate(new_width)
        self._width = new_width

    def get_height(self):
        return self._height

    def set_height(self, new_height):
        self._validate(new_height)
        self._height = new_height


```

```python
rect = Rectangle(10, 5)
print("Ширина:", rect.get_width())      
print("Высота:", rect.get_height())     
print("Периметр:", rect.perimeter())   
print("Площадь:", rect.area())          

 
```

```python
rect.set_width(20)
rect.set_height(15)
print("Новая ширина:", rect.get_width())    
print("Новая высота:", rect.get_height())  
print("Новый периметр:", rect.perimeter())  
print("Новая площадь:", rect.area())       
```

```python
class Temperature:
    def __init__(self, celsius=0):
        self._validate(celsius)
        self._celsius = celsius

    def _validate(self, value):
        if not isinstance(value, (int, float)) or not (100 <= value <= 200):
            raise ValueError("Не применимо")

    def to_fahrenheit(self):
        return self._celsius * 9 / 5 + 32

    def to_kelvin(self):
        return self._celsius + 273.15

    def get_celsius(self):
        return self._celsius

    def set_celsius(self, new_celsius):
        self._validate(new_celsius)
        self._celsius = new_celsius


```

```python
temp = Temperature(150)
print("Цельсий:", temp.get_celsius())        
print("Фаренгейт:", temp.to_fahrenheit())     
print("Кельвин:", temp.to_kelvin())           

```

```python
temp.set_celsius(180)
print("Новая Цельсий:", temp.get_celsius())   
print("Новый Фаренгейт:", temp.to_fahrenheit())
print("Новый Кельвин:", temp.to_kelvin())
```
