```python
import numpy as np
import pandas as pd


df = pd.read_csv("tips.csv")

# ==============================
# МЕТРИКА 1: Общий размер чаевых по дням недели
# ==============================


days = df['day'].unique()


day_tips = {}

for d in days:
    total_tips = df[df['day'] == d]['tip'].sum()
    day_tips[d] = total_tips
    print(f"{d} - {total_tips:.2f}")
```

```python
# ==============================
# МЕТРИКА 2: Процентное соотношение чаевых по дням недели
# ==============================


total_sum = df['tip'].sum()

for d in day_tips:
    percent = (day_tips[d] / total_sum) * 100
    print(f"{d} - {percent:.2f}%")
```

```python
# ==============================
# МЕТРИКА 3: Три компании посетителей с наибольшими суммами чаевых
# ==============================


sizes = df['size'].unique()


size_tips = {}

for s in sizes:
    total = df[df['size'] == s]['tip'].sum()
    size_tips[s] = total

sorted_sizes = sorted(size_tips.items(), key=lambda x: x[1], reverse=True)


top_3 = sorted_sizes[:3]

for group, total in top_3:
    print(f"{group} человек(а) - {total:.2f}")
```
