```python
num1 = int(input('введите первое число: '))
num2 = int(input('введите второе число: '))

if num1 > num2:
    print("Диапазон неверен")
else:
    for i in range(num1, num2 + 1):
        print(i, end=' ')
```

```python
num = int(input("Введите число: "))

for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

```python
num = input("Введите строку из цифр: ")

for i in num:
    if int(i) % 2 != 0:
        print("Not clear")
        break
else:
    print("All clear")
```

```python
s = input("Введите строку из цифр: ")

total = 0         
count = 0          

for ch in s:
    total += int(ch)
    count += 1

average = total // count
first_last_sum = int(s[0]) + int(s[-1])

print("Сумма всех чисел -", total)
print("Количество всех чисел -", count)
print("Средние арифметическое всех чисел -", average)
print("Сумма первой и последний цифры -", first_last_sum)
```

```python
num_1 = int(input("Введите первое число: "))   #в этой задаче немного использовал chatGPT
num_2 = int(input("Введите второе число: "))

for n in range(num_1, num_2 + 1):
    count_div = 0
    for d in range(1, n + 1):
        if n % d == 0:
            count_div += 1
    print(str(n) + "+" * count_div)
```
