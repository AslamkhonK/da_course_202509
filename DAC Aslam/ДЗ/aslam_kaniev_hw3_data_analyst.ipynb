```python
from fractions import Fraction
```

```python
# Задача 1:
# В колоде 52 карты, из них 4 туза.
# Мы поочередно вытаскиваем 2 карты без возврата.
# Нужно найти вероятность того, что обе карты окажутся тузами.


p_two_aces = Fraction(4, 52) * Fraction(3, 51)
print("Вероятность двух тузов:", float(p_two_aces))
```

```python
# Задача 2:
# Друг просыпает поздно в 80% случаев. Если просыпает, то опаздывает с вероятностью 70%.
# Если просыпается вовремя (20% случаев), то опаздывает с вероятностью 10%.
# Найти общую вероятность опоздания.


P_L = Fraction(4, 5)           
P_notL = Fraction(1, 5)        
P_A_given_L = Fraction(7, 10)  
P_A_given_notL = Fraction(1, 10)  

P_A = P_A_given_L * P_L + P_A_given_notL * P_notL
print("Вероятность опоздать:", float(P_A))
```

```python
# Задача 3:
# В мешке 3 белых, 4 зелёных и 5 красных шаров (всего 12).
# Нужно вытащить два шара без возврата.
# Найти вероятность, что один шар зелёный, а другой красный — порядок не важен.

from fractions import Fraction

p_G_then_R = Fraction(4, 12) * Fraction(5, 11)  # зелёный → красный
p_R_then_G = Fraction(5, 12) * Fraction(4, 11)  # красный → зелёный
p_GR = p_G_then_R + p_R_then_G

print("Вероятность зелёный+красный:", float(p_GR))
```

```python
# Задача 4:
# 98% упаковок — с добавками, 2% — без.
# Испытуемый отвечает "Да":
#   - в 99% случаев, если добавки есть,
#   - в 2.5% случаев, если добавок нет.
# Нужно найти вероятность, что если человек ответил "Нет",
# то упаковка действительно без добавок (формула Байеса).


P_A = Fraction(98, 100)     
P_notA = Fraction(2, 100)   

P_No_given_A = Fraction(1, 100)          
P_No_given_notA = Fraction(975, 1000)   

P_No = P_No_given_A * P_A + P_No_given_notA * P_notA  
P_notA_given_No = P_No_given_notA * P_notA / P_No

print("P(нет добавок | 'Нет'):", float(P_notA_given_No))
```

```python
# Задача 5:
# 3% двигателей перегреваются.
# При перегреве индикатор включается в 98% случаев.
# При отсутствии перегрева индикатор включается в 1% случаев.
# Нужно:
#  1) Найти общую вероятность того, что индикатор включён.
#  2) Найти вероятность, что двигатель перегрелся, если индикатор включён.

from fractions import Fraction

P_H = Fraction(3, 100)         
P_notH = Fraction(97, 100)      

P_I_given_H = Fraction(98, 100)    
P_I_given_notH = Fraction(1, 100)   

# 1. Вероятность включения индикатора
P_I = P_I_given_H * P_H + P_I_given_notH * P_notH

# 2. Вероятность перегрева при включенном индикаторе 
P_H_given_I = P_I_given_H * P_H / P_I

print("P(индикатор включён):", float(P_I))
print("P(перегрев | индикатор включён):", float(P_H_given_I))
```
