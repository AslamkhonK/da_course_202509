```python
data = {
  'Продукт': ['Молоко', 'Хлеб', 'Яблоки', 'Мясо', 'Рис', 'Сахар', 'Картофель', 'Сыр', 'Макароны',
              'Йогурт', 'Колбаса', 'Масло', 'Чай', 'Кофе', 'Курица'],
  'Категория': ['Молочная', 'Выпечка', np.nan, 'Мясо', np.nan, 'Сладости', 'Овощи', np.nan, 'Бакалея', 'Молочная',
                'Колбасы', 'Молочная', 'Напитки', 'Напитки', np.nan],
  'Цена': [1.5, 0.8, 2.3, 5.0, 1.2, 0.9, 1.0, 4.5, 1.1, 0.7, 3.5, 2.0, 1.8, 3.2, 4.0],
  'Количество': [50, np.nan, 30, 20, np.nan, 40, 60, 15, 70, 25, 35, np.nan, 45, 30, 20]
}


df_inventory = pd.DataFrame(data)


print("Типы данных в таблице:")
print(df_inventory.dtypes)



numeric_cols = df_inventory.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    df_inventory[col] = df_inventory[col].fillna(df_inventory[col].mean())

cat_cols = df_inventory.select_dtypes(include=['object']).columns
for col in cat_cols:
    df_inventory[col] = df_inventory[col].fillna(df_inventory[col].mode()[0])


print("\nТаблица после заполнения пропусков:")
print(df_inventory)


df_inventory.to_csv("inventory_filled.csv", index=False)
```
