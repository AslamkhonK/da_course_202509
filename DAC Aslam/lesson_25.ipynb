```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("mpg")
```

```python
mean = df['mpg'].mean()
std_err = df['mpg'].std() /np.sqrt(df['mpg'].size)
ci = 1.96 * std_err
```

```python
sns.boxplot(df, y='mpg', hue='origin')

plt.errorbar(x= 0, y= mean, yerr=ci)   
```
