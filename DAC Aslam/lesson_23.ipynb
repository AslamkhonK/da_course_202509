```python
import pandas as pd
import numpy as np
```

```python
def count_all_rows(dct_df: dict[str, pd.DataFrame]):
    couter = 0

    for key in dct_df:
        print(f"{key} - {dct_df[key].shape[0]} rows")
        couter += dct_df[key].shape[0]

    return couter
```

```python
def axe_analitic(column):
    size = column.shape[0]
    null = column.isnull()

    return f"{round(null.sum() * 100 / size, 2)}%"
```

```python
table_columns = {'titanik': ['class', 'age'], 'mpg': ['name', 'origin', 'mpg']}
```

```python
def check_columns(df_dict, table_columns):
    for k, v in table_columns.items():
        print(f"Таблица - {k}")

        for col in v:
            Na_per = axe_analitic(df_dict[k][col])
            print(f"Столбец - {col} - {Na_per} пропущенных")
        
        print('-' * 40)
```

```python
check_columns(df_dict=df_dict, table_columns=table_columns)
```

```python
def check_columns2(df_dict, **kwargs):
    print(kwargs)
    for k, v in kwargs.items():
        print(f"Таблица - {k}")

        for col in v:
            Na_per = axe_analitic(df_dict[k][col])
            print(f"Столбец - {col} - {Na_per} пропущенных")
        
        print('-' * 40)
```

```python
check_columns2(df_dict=df_dict, titanik=['class', 'age'], mpg=['name', 'origin', 'mpg'])
```
