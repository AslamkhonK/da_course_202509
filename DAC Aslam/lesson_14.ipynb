GUN

```python
class Gun():
    def __init__(self, ammo=7):
        self.ammo = ammo
        self.max_ammo = ammo
    def shoot(self):
        if self.ammo > 0:
            self.ammo -= 1
            print("Bang!")
        else:
            print("No ammo left!")
    def reload(self, bullets):
        self.ammo = self.max_ammo
        print(f"Reloaded to {self.ammo} bullets")
```

```python
pistol = Gun(10)
```

```python
for i in range(7):
    pistol.shoot()
```

```python
pistol.reload(7)
```
