```python
import random

secret_number = random.randint(1, 30)

while True:
    guess = int(input("Угадайте число от 1 до 30: "))

    if guess > secret_number:
        print("Слишком много, попробуйте еще раз.")
    elif guess < secret_number:
        print("Слишком мало, попробуйте еще раз.")
    else:
        print("Вы угадали, поздравляем!")
        break 
```

```python
import random


secret_number = random.randint(1, 30)

print("Угадайте число от 1 до 30. У вас есть 5 попыток!")

for attempt in range(1, 6):
    guess = int(input(f"Попытка {attempt}: введите число: "))

    if guess > secret_number:
        print("Слишком много, попробуйте еще раз.")
    elif guess < secret_number:
        print("Слишком мало, попробуйте еще раз.")
    else:
        print(f"Вы угадали! Поздравляем! Число было {secret_number}.")
        break
else:
    print(f"Попытки закончились Загаданное число было {secret_number}.")
```

```python
import random as r

rnd = r.randint(1, 30)
cnt_enters = 0

while (player_num := int(input('Какое число было загадано? | '))) != rnd:
    
    if player_num == 0:
        print("### Ошибка ввода ###")
        continue

    if player_num > rnd:
        print(f"Загаданное число меньше чем {player_num}")
    else:
        print(f"Загаданное число больше чем {player_num}")

    cnt_enters += 1
    if cnt_enters == 5:
        print(f"Неудача, затрачено слишком много попыток, вы не угадали(")
        print(f"Загаданное число было - {rnd}")
        break
else:
    print(f"Поздравляю, вы угадали число с {cnt_enters+1} попытки!")
    print(f"Загаданное число это - {rnd}")
```
