```python

num = int(input("Введите первое число: "))
nom = int(input("Введите второе число: "))
op = input("Введите операцию (+, -, *, /): ")

if op == '+':
    print(num + nom)
elif op == '-':
    print(num - nom)
elif op == '*':
    print(num * nom)
elif op == '/':
    if num == 0:
        print("На ноль делить нельзя!")
    else:
        print(num / nom)
else:
    print("Неверная операция")
```
