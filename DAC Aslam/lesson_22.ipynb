```python
df = pd.DataFrame({
    'name': ['Emily', 'Emily', 'James', 'James'],
    'subject': ['Literature', 'Mathematics', 'Literature', 'Mathematics'],
    'max_score': [120, 150, 120, 150],
    'curr_score': [98, 78, 80, 131]
    })
```

```python
df
```

```python
def return_in_per(mx_sxore, curr_score):
    r = (curr_score / mx_sxore) * 100
    return f"{round(r, 2)}%"
```

```python
return_in_per(150, 78)
```

```python
df[['max_score', 'curr_score']].apply(lambda var_df: return_in_per(var_df['max_score'], var_df['curr_score']), axis=1)
```
