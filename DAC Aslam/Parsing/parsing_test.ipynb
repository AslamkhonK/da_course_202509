```python
import requests
from bs4 import BeautifulSoup
```

```python
def create_file(url, page_num):
    page = requests.get(url + f"?page={page_num}")
    with open(f"page{page_num}.html", "w", encoding='utf-8') as html_file:
        html_file.write(page.text)
```

```python
create_file(
    url="https://somon.tj/transport/legkovyie-avtomobili/dushanbe/",
    page_num=20
)
```

```python
with open('page20.html', 'r', encoding='utf-8') as file:
    html_file = file.read()

bs = BeautifulSoup(html_file, 'lxml')

price_tags = bs.find_all('a', class_="advert__content-price _not-title")
title_tags = bs.find_all('a', class_="advert__content-title")

cars = []
```

```python
for title_tag, price_tag in zip(title_tags, price_tags):
    parts = title_tag.text.strip().split(', ')
    
    car_model = parts[0]                     
    year = parts[1] if len(parts) > 1 else "-"  
    price = price_tag.text.split('c.')[0].strip().replace(' ', '')  
    
    cars.append([car_model, year, price])
```

```python
print("{:<30} {:<12} {:<10}".format("Модель", "Год", "Цена (TJS)"))
print("-" * 55)
for car in cars:
    print("{:<30} {:<12} {:<10}".format(car[0], car[1], car[2])) 
```
