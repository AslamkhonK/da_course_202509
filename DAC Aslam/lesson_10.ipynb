```python
def my_map(func, lst):
    I = []
    for i in lst:
        I.append(func(i))
    return I
```

```python
my_map(func=abs, lst=[1, -3, -2, -4, 6, 5 ])
```

```python
def make_letter_pairs(text: str):
    
    letters = filter(str.isalpha, text)

    pairs = map(lambda ch: (ch.upper(), ch.lower()), letters)
    return list(pairs)

text = "Привет, World! 123"
result = make_letter_pairs(text)
print(result)
```
