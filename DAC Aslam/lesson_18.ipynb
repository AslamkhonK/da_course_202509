```python
class ParentCalss:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
     return f"Name: {self.name}, Age: {self.age}"
    def sey_hallo(self):
        return f"Hallo, I am {self.name}"
class ChildClass(ParentCalss):
    def __init__(self, name, age, surname, salary):
        super().__init__(name, age)
        self.surname = surname
        self.salary = salary

    
```

```python
child = ChildClass('John', 30, 'Doe', salary=50000)
child
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Cat(Animal):
    def sound(self):
        return "Meow!"

class Dog(Animal):
    def sound(self):
        return "Woof!"    
        
```

```python
cat = Cat("Whiskers")
```

```python
dog = Dog("Buddy")
```

```python
for i in (cat, dog):
    print(i.sound())
```
