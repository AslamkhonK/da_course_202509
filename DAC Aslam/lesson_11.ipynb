```python
a = int(input('First number: '))
b = int(input('Second number: '))
c = int(input('Third number: '))
d = int(input('Fourth number: '))

print(f'–í–≤–µ–¥–µ–Ω—ã —á–∏—Å–ª–∞: {a}, {b}, {c}, {d}')

# –ú–∞–∫—Å–∏–º—É–º
max_number = max(a, b, c, d)
print(f'–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ: {max_number}')

# –ß—ë—Ç–Ω—ã–µ —á–∏—Å–ª–∞
even_numbers = []
for num in [a, b, c, d]:
    if num % 2 == 0:
        even_numbers.append(num)

if even_numbers:
    print(f'–ß—ë—Ç–Ω—ã–µ —á–∏—Å–ª–∞: {even_numbers}')
else:
    print('–ß—ë—Ç–Ω—ã—Ö —á–∏—Å–µ–ª –Ω–µ—Ç')

# –ö–≤–∞–¥—Ä–∞—Ç—ã
squares = [a**2, b**2, c**2, d**2]
print(f'–ö–≤–∞–¥—Ä–∞—Ç—ã —á–∏—Å–µ–ª: {squares}')

# –°—É–º–º–∞
total = a + b + c + d
print(f'–°—É–º–º–∞ –≤—Å–µ—Ö —á–∏—Å–µ–ª: {total}')
```
